```python
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import pydotplus
from IPython.display import Image 
```

```python
cancer=load_breast_cancer()
cancer
```

```python
clf=DecisionTreeClassifier().fit(cancer.data,cancer.target)

dot_data = export_graphviz(clf, out_file=None, filled=True, rounded=True,
                                feature_names=cancer.feature_names,  
                                class_names=['mean symmetry','smoothness'])
graph=pydotplus.graph_from_dot_data(dot_data)
image=graph.create_png()
graph.write_png("imagen.png")
Image(filename="imagen.png")
```
